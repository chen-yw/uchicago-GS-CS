```python
gdgx_gs_2025_cleaned = gdgx_gs_2025[['id','name','address','plongcj02','platgcj02','tel','pname','cityname','adname','大类','中类','小类','updatetime']].copy()
gdgx_gs_2025_cleaned
```

```python
gdgx_cs_2025_cleaned = gdgx_cs_2025[['id','name','address','plongcj02','platgcj02','tel','pname','cityname','adname','大类','中类','小类','updatetime']].copy()
gdgx_cs_2025_cleaned
```

```python
import json
pname_translation_dict = json.load(open(f"{data_base_dir}/variable_dict/all_pname_translation_dict.json", 'r', encoding='utf-8'))
cityname_translation_dict = json.load(open(f"{data_base_dir}/variable_dict/all_cityname_translation_dict.json", 'r', encoding='utf-8'))
print(pname_translation_dict)
print(cityname_translation_dict)
```

```python
# Translate pname and cityname to English for GS data
gdgx_gs_2025_cleaned['pname_EN'] = gdgx_gs_2025_cleaned['pname'].map(pname_translation_dict)
gdgx_gs_2025_cleaned['cityname_EN'] = gdgx_gs_2025_cleaned['cityname'].map(cityname_translation_dict)

# Translate pname and cityname to English for CS data  
gdgx_cs_2025_cleaned['pname_EN'] = gdgx_cs_2025_cleaned['pname'].map(pname_translation_dict)
gdgx_cs_2025_cleaned['cityname_EN'] = gdgx_cs_2025_cleaned['cityname'].map(cityname_translation_dict)

print("Translation completed")
```

```python
gdgx_gs_2025_cleaned
```

```python
# Rename coordinate columns for both GS and CS data
gdgx_gs_2025_cleaned.rename(columns={'plongcj02': 'wgs84_x', 'platgcj02': 'wgs84_y'}, inplace=True)
gdgx_cs_2025_cleaned.rename(columns={'plongcj02': 'wgs84_x', 'platgcj02': 'wgs84_y'}, inplace=True)

print("Column names updated")
```

```python
# Check column consistency
print("GS 2025 columns:", list(GS_data_set[2025].columns))
print("GDGX GS 2025 columns:", list(gdgx_gs_2025_cleaned.columns))
print("CS 2025 columns:", list(CS_data_set[2025].columns))
print("GDGX CS 2025 columns:", list(gdgx_cs_2025_cleaned.columns))

# Merge GS datasets
merged_gs_2025 = pd.concat([GS_data_set[2025], gdgx_gs_2025_cleaned], ignore_index=True)
print(f"Merged GS 2025 length: {len(merged_gs_2025)}")

# Merge CS datasets
merged_cs_2025 = pd.concat([CS_data_set[2025], gdgx_cs_2025_cleaned], ignore_index=True)
print(f"Merged CS 2025 length: {len(merged_cs_2025)}")

print("Data merge completed")
```

```python
# Fix updatetime column data type for parquet compatibility
merged_gs_2025['updatetime'] = merged_gs_2025['updatetime'].astype(str)
merged_cs_2025['updatetime'] = merged_cs_2025['updatetime'].astype(str)

print("Data types fixed for parquet export")
```

```python
merged_gs_2025.to_parquet(f"{data_GS_dir}/GS_data_2025.parquet")
merged_cs_2025.to_parquet(f"{data_CS_dir}/CS_data_2025.parquet")
```
