```python
matching_columns = ['pname','cityname','adname','address','name']
```

```python
import numpy as np

def get_remove_mask_by_duplicate_and_paused(data_set, matching_columns, distance_func, threshold=500):
    """
    For each year, group by matching_columns, check duplicates by distance, 
    and mark all as remove if any paused, else keep one randomly.
    Returns a dict: {year: mask (True means to remove)}
    """
    remove_mask_dict = {}
    for year, df in data_set.items():
        print(f"Processing year: {year}")
        mask = np.zeros(len(df), dtype=bool)
        mask[df['paused'] == 1] = True
        grouped = df.groupby(matching_columns)
        for _, group in grouped:
            idxs = group.index.tolist()
            coords = group[['wgs84_x', 'wgs84_y']].values
            n = len(idxs)
            to_remove = set()
            # Check pairwise distances
            for i in range(n):
                for j in range(i+1, n):
                    dist = distance_func(coords[i][0], coords[i][1], coords[j][0], coords[j][1])
                    if dist < threshold:
                        # If any paused, mark all as remove
                        if group.iloc[[i, j]]['paused'].any():
                            to_remove.add(idxs[i])
                            to_remove.add(idxs[j])
                        else:
                            to_remove.add(idxs[j])
            for idx in to_remove:
                mask[idx] = True
                        
        remove_mask_dict[year] = mask
    return remove_mask_dict
```

```python
def yearly_anchor_matching_stats(data_set, anchor_matching_columns):
    years = sorted(data_set.keys())
    result = {}
    for i in range(len(years) - 1):
        year1, year2 = years[i], years[i+1]
        df1 = data_set[year1]
        df2 = data_set[year2]
        df2_copy = df2.copy()
        # Filter out entries with empty or meaningless address
        df1 = df1[(df1['address'].str.strip() != '') & (df1['address'].str.strip() != '[]')]
        df2 = df2[(df2['address'].str.strip() != '') & (df2['address'].str.strip() != '[]')]
        # Merge on anchor columns
        merged = df1.merge(df2, on=anchor_matching_columns, suffixes=('_y1', '_y2'))
        # Group by cityname
        city_stats = merged.groupby('cityname').size().to_dict()
        total_city_samples = {city: len(df2_copy[df2_copy['cityname'] == city]) for city in city_stats}
        match_ratio = {city: city_stats[city] / total_city_samples[city] if total_city_samples[city] > 0 else 0 for city in city_stats}
        result[(year1, year2)] = {
            'total_matched': len(merged),
            'total_samples_next_year': len(df2),
            'match_ratio': len(merged) / len(df2) if len(df2) > 0 else 0,
            'city_stats': {
                city: {
                    'matched': city_stats[city],
                    'total_next_year': total_city_samples[city],
                    'ratio': match_ratio[city]
                } for city in city_stats
            }
        }
    return result
```

```python
# Example usage:
anchor_matching_columns = ['pname','cityname','adname','address','name']
stats = yearly_anchor_matching_stats(GS_data_set, anchor_matching_columns)
stats
```

```python
def plot_city_match_distribution(stats, label):
    year_pairs = list(stats.keys())
    n = len(year_pairs)
    fig, axs = plt.subplots(n, 2, figsize=(12, 4*n), sharey='col')
    for i, year_pair in enumerate(year_pairs):
        city_ratios = [v['ratio'] for v in stats[year_pair]['city_stats'].values()]
        city_counts = [v['matched'] for v in stats[year_pair]['city_stats'].values()]
        axs[i, 0].hist(city_ratios, bins=20, color='skyblue', edgecolor='black')
        axs[i, 0].set_title(f'{label} {year_pair[0]}-{year_pair[1]} City Match Ratio')
        axs[i, 0].set_xlabel('Match Ratio')
        axs[i, 0].set_ylabel('City Count')
        axs[i, 1].hist(city_counts, bins=10, color='salmon', edgecolor='black')
        axs[i, 1].set_title(f'{label} {year_pair[0]}-{year_pair[1]} City Match Count')
        axs[i, 1].set_xlabel('Matched Count')
        axs[i, 1].set_ylabel('City Count')
    plt.tight_layout()
    plt.show()
```
