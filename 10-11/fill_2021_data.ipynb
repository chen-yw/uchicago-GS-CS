```python
def fill_specific_year_x_y_data(data_set, year, x_col, y_col, matching_cols=[]):
    total_row = len(data_set[year])
    empty_row = 0
    failed_fill_row = 0
    that_year_data = data_set[year]
    prev_year_data = data_set[year-1]
    next_year_data = data_set[year+1]
    print(f"Filling year {year} data, total rows: {total_row}")
    for idx, row in that_year_data.iterrows():
        if pd.isna(row[x_col]) or pd.isna(row[y_col]):
            empty_row += 1
            matched_prev = prev_year_data
            matched_next = next_year_data
            for col in matching_cols:
                matched_prev = matched_prev[matched_prev[col] == row[col]]
                matched_next = matched_next[matched_next[col] == row[col]]
            if len(matched_prev) > 0:
                that_year_data.at[idx, x_col] = matched_prev.iloc[0][x_col]
                that_year_data.at[idx, y_col] = matched_prev.iloc[0][y_col]
            elif len(matched_next) > 0:
                that_year_data.at[idx, x_col] = matched_next.iloc[0][x_col]
                that_year_data.at[idx, y_col] = matched_next.iloc[0][y_col]
            else:
                failed_fill_row += 1
                print(f"Failed to fill row {idx}, name: {row['name']}")
    print(f"Total rows: {total_row}, Empty rows: {empty_row}, Failed to fill rows: {failed_fill_row}")
    return that_year_data
```

```python
def fill_specific_year_x_y_data_efficient(data_set, year, x_col, y_col, matching_cols=[]):
    """
    Parameters:
    - data_set: Dictionary containing data for different years
    - year: Target year to fill missing data
    - x_col: Column name for x coordinate (e.g., 'wgs84_x')
    - y_col: Column name for y coordinate (e.g., 'wgs84_y')
    - matching_cols: List of columns to use for matching records across years
    
    Returns:
    - DataFrame with filled coordinates
    """
    
    # Get data for the target year and adjacent years
    target_data = data_set[year].copy()
    prev_year_data = data_set[year-1].copy() if year-1 in data_set else None
    next_year_data = data_set[year+1].copy() if year+1 in data_set else None
    
    total_rows = len(target_data)
    # Count rows with missing coordinates
    missing_mask = target_data[x_col].isna() | target_data[y_col].isna()
    empty_rows = missing_mask.sum()
    
    print(f"Filling year {year} data, total rows: {total_rows}, rows with missing coordinates: {empty_rows}")
    
    if empty_rows == 0:
        print("No missing coordinates found, returning original data")
        return target_data
    
    # Add a temporary index to preserve original row order
    target_data = target_data.reset_index(drop=False).rename(columns={'index': 'original_index'})
    
    # First, try to fill from previous year data
    if prev_year_data is not None and len(prev_year_data) > 0:
        # Prepare previous year data for merging
        prev_coords = prev_year_data[matching_cols + [x_col, y_col]].copy()
        prev_coords = prev_coords.dropna(subset=[x_col, y_col])  # Only keep rows with valid coordinates
        prev_coords = prev_coords.rename(columns={x_col: f'{x_col}_prev', y_col: f'{y_col}_prev'})
        
        # Merge with previous year data
        target_with_prev = target_data.merge(prev_coords, on=matching_cols, how='left')
        
        # Fill missing coordinates with previous year data
        prev_fill_mask = (target_with_prev[x_col].isna() | target_with_prev[y_col].isna()) & \
                        target_with_prev[f'{x_col}_prev'].notna() & target_with_prev[f'{y_col}_prev'].notna()
        
        target_with_prev.loc[prev_fill_mask, x_col] = target_with_prev.loc[prev_fill_mask, f'{x_col}_prev']
        target_with_prev.loc[prev_fill_mask, y_col] = target_with_prev.loc[prev_fill_mask, f'{y_col}_prev']
        
        filled_from_prev = prev_fill_mask.sum()
        print(f"Filled {filled_from_prev} rows from previous year ({year-1})")
        
        # Clean up temporary columns
        target_data = target_with_prev.drop(columns=[f'{x_col}_prev', f'{y_col}_prev'])
    
    # Then, try to fill remaining missing data from next year
    if next_year_data is not None and len(next_year_data) > 0:
        # Check if there are still missing coordinates
        remaining_missing_mask = target_data[x_col].isna() | target_data[y_col].isna()
        remaining_missing = remaining_missing_mask.sum()
        
        if remaining_missing > 0:
            # Prepare next year data for merging
            next_coords = next_year_data[matching_cols + [x_col, y_col]].copy()
            next_coords = next_coords.dropna(subset=[x_col, y_col])  # Only keep rows with valid coordinates
            next_coords = next_coords.rename(columns={x_col: f'{x_col}_next', y_col: f'{y_col}_next'})
            
            # Merge with next year data
            target_with_next = target_data.merge(next_coords, on=matching_cols, how='left')
            
            # Fill remaining missing coordinates with next year data
            next_fill_mask = (target_with_next[x_col].isna() | target_with_next[y_col].isna()) & \
                            target_with_next[f'{x_col}_next'].notna() & target_with_next[f'{y_col}_next'].notna()
            
            target_with_next.loc[next_fill_mask, x_col] = target_with_next.loc[next_fill_mask, f'{x_col}_next']
            target_with_next.loc[next_fill_mask, y_col] = target_with_next.loc[next_fill_mask, f'{y_col}_next']
            
            filled_from_next = next_fill_mask.sum()
            print(f"Filled {filled_from_next} rows from next year ({year+1})")
            
            # Clean up temporary columns
            target_data = target_with_next.drop(columns=[f'{x_col}_next', f'{y_col}_next'])
    
    # Calculate final statistics
    final_missing_mask = target_data[x_col].isna() | target_data[y_col].isna()
    final_missing = final_missing_mask.sum()
    successfully_filled = empty_rows - final_missing
    
    # print(f"Summary: Total rows: {total_rows}, Originally missing: {empty_rows}, Successfully filled: {successfully_filled}, Still missing: {final_missing}")
    
    # Restore original index order and remove temporary index
    target_data = target_data.sort_values('original_index').drop(columns=['original_index']).reset_index(drop=True)
    
    return target_data
```

```python
def check_how_many_merge_with_different_cols(data_set, year1, year2, cols):
    df_year1 = data_set[year1]
    df_year2 = data_set[year2]
    merged = df_year1.merge(df_year2, on=cols, suffixes=('_year1', '_year2'))
    print(f"Year {year1} rows: {len(df_year1)}, Year {year2} rows: {len(df_year2)}")
    print(f"Merging year {year1} and year {year2} on columns {cols}, resulting rows: {len(merged)}")
    return merged
```
