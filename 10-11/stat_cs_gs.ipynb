```python
# Method 1: Check for duplicate values in specified column combinations
def check_duplicates_by_columns(df, columns, year=None):
    """
    Check if there are duplicate values in specified column combinations
    df: DataFrame
    columns: List of column names to check
    year: Year (for printing information)
    """
    # Check if these column combinations have duplicates
    duplicates = df.duplicated(subset=columns, keep=False)
    duplicate_count = duplicates.sum()
    
    if year:
        print(f"Year {year}:")
    print(f"  Total rows: {len(df)}  Duplicate rows: {duplicate_count}")
    
    # If there are duplicates, show the duplicate records
    if duplicate_count > 0:
        show_columns = columns + ["name","pname","cityname","adname"]
        # print(f"  Duplicate records:")
        duplicate_records = df[duplicates].sort_values(by=columns)
        # print(duplicate_records[show_columns].head(10))  # Only show first 10 records
    
    return duplicate_count

```

```python
location_columns = ['wgs84_x', 'wgs84_y']

for key in CS_data_set:
    data_single_year = CS_data_set[key]
    check_duplicates_by_columns(data_single_year, location_columns, key)
```
