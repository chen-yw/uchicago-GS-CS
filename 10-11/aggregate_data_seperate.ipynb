# This file is used to aggregate CS and GS data seperately
- By year and city, showing total count for each year and the year-over-year change.

```python
import pandas as pd
```

```python
CS_data_set = {}
base_path = "C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/raw/全国加油站加气站数据"
CS_dir = f"{base_path}/充电站"
for year in range(2015,2026):
    CS_data_path = f"{CS_dir}/全国加油站加气站数据{year}_充电站.csv"
    CS_data_single_year = pd.read_csv(CS_data_path)
    CS_data_set[year] = CS_data_single_year
    print(year, len(CS_data_single_year))
```

```python
GS_data_set = {}
base_path = "C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/raw/全国加油站加气站数据"
GS_dir = f"{base_path}/加油站加气站"
for year in range(2013,2026):
    GS_data_path = f"{GS_dir}/全国加油站加气站数据{year}_加油站加气站.csv"
    GS_data_single_year = pd.read_csv(GS_data_path)
    GS_data_set[year] = GS_data_single_year
    print(year, len(GS_data_single_year))
```

```python
GS_data_set[2021]
```

```python
def get_year_counts_with_size(data,count_name):
    grouped = data.groupby(["pname","cityname"]).size().reset_index(name=count_name)
    return grouped

result = get_year_counts_with_size(GS_data_set[2021],"GS_count")
print(result)
result = get_year_counts_with_size(CS_data_set[2021],"CS_count")
print(result)
```

```python
def aggregate_data_by_year(data_set, count_name):
    all_years_data = []
    for year, data in data_set.items():
        yearly_counts = get_year_counts_with_size(data, count_name)
        yearly_counts['year'] = year
        all_years_data.append(yearly_counts)
    result = pd.concat(all_years_data, ignore_index=True)
    columns = ['year'] + [col for col in result.columns if col != 'year']
    return result[columns]
```

```python
GS_aggregated = aggregate_data_by_year(GS_data_set, "GS_count")
GS_aggregated
```

```python
CS_aggregated = aggregate_data_by_year(CS_data_set, "CS_count")
CS_aggregated
```

```python
def calculate_yearly_changes(aggregated_data, count_column):
    aggregated_data = aggregated_data.sort_values(by=['pname', 'cityname', 'year'])
    aggregated_data['prev_count'] = aggregated_data.groupby(['pname', 'cityname'])[count_column].shift(1)
    aggregated_data['yearly_change'] = aggregated_data[count_column] - aggregated_data['prev_count']
    aggregated_data.drop(columns=['prev_count'], inplace=True)  # 删除辅助列
    return aggregated_data
```
