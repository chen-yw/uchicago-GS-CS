```python
def check_how_many_merge_with_different_cols(data_set, year1, year2, cols):
    df_year1 = data_set[year1]
    df_year2 = data_set[year2]
    merged = df_year1.merge(df_year2, on=cols, suffixes=('_year1', '_year2'))
    print(f"Year {year1} rows: {len(df_year1)}, Year {year2} rows: {len(df_year2)}")
    print(f"Merging year {year1} and year {year2} on columns {cols}, resulting rows: {len(merged)}")
    return merged
```
