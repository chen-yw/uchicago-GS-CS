```python
base_dir = 'C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/intermediate/yiwei'
cs_chinese = pd.read_csv(f"{base_dir}/CS_aggregated_chinese.csv")
cs_chinese
```

```python
gs_chinese = pd.read_csv(f"{base_dir}/GS_aggregated_chinese.csv")
gs_chinese
```

```python
def get_unique_values(column_name):
    unique_cs = cs_chinese[column_name].unique()
    unique_gs = gs_chinese[column_name].unique()
    unique_values = list(set(list(unique_cs) + list(unique_gs)))
    return unique_values
```

```python
column_name = 'pname'
unique_pname_values = get_unique_values(column_name)
print(len(unique_pname_values))
print(unique_pname_values)
```

```python
column_name = 'cityname'
unique_cityname_values = get_unique_values(column_name)
print(len(unique_cityname_values))
print(unique_cityname_values)
```

```python
def generate_translation_dict(unique_values):
    translation_dict = {}
    for value in unique_values:
        try:
            translated_value = translator.translate(value)
            translation_dict[value] = translated_value
        except Exception as e:
            print(f"Error translating {value}: {e}")
            translation_dict[value] = None
    return translation_dict
```

```python
pname_dict = generate_translation_dict(unique_pname_values)
pname_dict
```

```python
def generate_translation_dict_batch(unique_values):
    translation_dict = {}
    try:
        # 过滤掉空值和None值
        valid_values = [v for v in unique_values if v is not None and str(v).strip() != '']
        
        # 批量翻译
        translated_values = translator.translate_batch(valid_values)
        
        # 创建字典映射
        for original, translated in zip(valid_values, translated_values):
            translation_dict[original] = translated
            
    except Exception as e:
        print(f"Batch translation failed: {e}")
        print("Falling back to individual translation...")
        
        # 如果批量翻译失败，回退到逐个翻译
        for value in unique_values:
            try:
                translated_value = translator.translate(value)
                translation_dict[value] = translated_value
            except Exception as e:
                print(f"Error translating {value}: {e}")
                translation_dict[value] = None
                
    return translation_dict
```

```python
cityname_dict = generate_translation_dict_batch(unique_cityname_values)
cityname_dict
```

```python
cs_chinese['pname_en'] = cs_chinese['pname'].map(pname_dict)
cs_chinese['cityname_en'] = cs_chinese['cityname'].map(cityname_dict)

gs_chinese['pname_en'] = gs_chinese['pname'].map(pname_dict)
gs_chinese['cityname_en'] = gs_chinese['cityname'].map(cityname_dict)
```

```python
columns = ['year','pname','pname_en','cityname','cityname_en','CS_count','yearly_change']
cs_final = cs_chinese[columns]
cs_final
```

```python
columns = ['year','pname','pname_en','cityname','cityname_en','GS_count','yearly_change']
gs_final = gs_chinese[columns]
gs_final
```

```python
cs_final.to_csv(f"{base_dir}/CS_aggregated_final.csv", index=False)
gs_final.to_csv(f"{base_dir}/GS_aggregated_final.csv", index=False)
```
