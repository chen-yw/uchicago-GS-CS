```python
fill_columns = ['wgs84_x', 'wgs84_y']

def fill_data_by_match(old_df, new_df, match_columns, fill_columns):
    result_df = old_df.copy()
    
    merged = pd.merge(old_df, new_df[match_columns + fill_columns], 
                     on=match_columns, how='left', suffixes=('', '_new'))
    
    for col in fill_columns:
        new_col = f"{col}_new"
        if new_col in merged.columns:
            result_df[col] = merged[col].combine_first(merged[new_col])
    
    return result_df

updated_GS_2021 = fill_data_by_match(GS_data_set[2021], new_2021_data_GS,
                                    match_columns, ['wgs84_x', 'wgs84_y'])
updated_CS_2021 = fill_data_by_match(CS_data_set[2021], new_2021_data_CS,
                                    match_columns, ['wgs84_x', 'wgs84_y'])
```
