```python
import pandas as pd
from deep_translator import GoogleTranslator
```

```python
base_dir = "C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/intermediate/yiwei/intermidiate"
CS_data_dir = base_dir + "/CS_cleaned/"
GS_data_dir = base_dir + "/GS_cleaned/"
```

```python
CS_data_set = {}
GS_data_set = {}
for i in range(2013,2026):
    GS_data_set[i] = pd.read_parquet(GS_data_dir + f"{i}.parquet")
    print(f"GS data for year {i} loaded.Length: {len(GS_data_set[i])}")
print()
for i in range(2015,2026):
    CS_data_set[i] = pd.read_parquet(CS_data_dir + f"{i}.parquet")
    print(f"CS data for year {i} loaded.Length: {len(CS_data_set[i])}")
```

```python
import sys
sys.path.append("C:/Users/31155/Dropbox/EV-GasDualNetwork/code/yiwei/")
from utils.get_unique_names import get_unique_names
```

```python
GS_pname_set , GS_cityname_set = get_unique_names(GS_data_set)
CS_pname_set , CS_cityname_set = get_unique_names(CS_data_set)
```

```python
print(GS_pname_set)
print(CS_pname_set)
print(GS_cityname_set)
print(CS_cityname_set)
```

```python
all_pname_set = sorted(list(set(GS_pname_set+CS_pname_set)))
all_cityname_set = sorted(list(set(GS_cityname_set+CS_cityname_set)))
print(len(all_pname_set),all_pname_set)
print(len(all_cityname_set),all_cityname_set)
```

```python
translator = GoogleTranslator(source='zh-CN', target='en')

all_pname_translation_dict = {}
all_cityname_translation_dict = {}
for name in all_pname_set:
    translation = translator.translate(name)
    all_pname_translation_dict[name] = translation
    print(f"Translated {name} to {translation}")

for name in all_cityname_set:
    translation = translator.translate(name)
    all_cityname_translation_dict[name] = translation
    print(f"Translated {name} to {translation}")
```

```python
print(all_pname_translation_dict)
print(all_cityname_translation_dict)
```

```python
def update_names_vectorized(data_dict, pname_mapping, cityname_mapping):
    """
    Update province and city names in datasets using vectorized operations
    
    Args:
        data_dict: Dictionary of dataframes by year
        pname_mapping: Dictionary mapping old pname -> correct pname
        cityname_mapping: Dictionary mapping old cityname -> correct cityname
    """
    for year, df in data_dict.items():
        print(f"Updating year {year}...")
        
        # Update pname using vectorized map operation
        if 'pname' in df.columns:
            df['pname_EN'] = df['pname'].map(pname_mapping).fillna(df['pname'])
        
        # Update cityname using vectorized map operation  
        if 'cityname' in df.columns:
            df['cityname_EN'] = df['cityname'].map(cityname_mapping).fillna(df['cityname'])
    
    print("All datasets updated successfully!")

# Extract mapping dictionaries from correction info
pname_mapping = {cn_name: en_name for cn_name, en_name in all_pname_translation_dict.items()}
cityname_mapping = {cn_name: en_name for cn_name, en_name in all_cityname_translation_dict.items()}

print(f"Created pname mapping for {len(pname_mapping)} entries")
print(f"Created cityname mapping for {len(cityname_mapping)} entries")

# Update both CS and GS datasets
print("Updating CS datasets...")
update_names_vectorized(CS_data_set, pname_mapping, cityname_mapping)

print("Updating GS datasets...")
update_names_vectorized(GS_data_set, pname_mapping, cityname_mapping)
```
