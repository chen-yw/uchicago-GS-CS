```python
import pandas as pd
```

```python
base_path = "C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/raw/全国加油站加气站数据"
output_base_path = "C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/intermediate/yiwei/raw"
```

```python
CS_dir = f"{base_path}/充电站"
CS_dir_output = f"{output_base_path}/充电站"
CS_data_set = {}
for year in range(2015,2026):
    CS_data_path = f"{CS_dir}/全国加油站加气站数据{year}_充电站.csv"
    CS_data_single_year = pd.read_csv(CS_data_path)
    CS_data_set[year] = CS_data_single_year
    CS_data_single_year.to_parquet(f"{CS_dir_output}/全国加油站加气站数据{year}_充电站.parquet")
    print(year, len(CS_data_single_year))
```

```python
GS_dir = f"{base_path}/加油站加气站"
GS_dir_output = f"{output_base_path}/加油站加气站"
GS_data_set = {}
for year in range(2013,2026):
    GS_data_path = f"{GS_dir}/全国加油站加气站数据{year}_加油站加气站.csv"
    GS_data_single_year = pd.read_csv(GS_data_path)
    GS_data_set[year] = GS_data_single_year
    GS_data_single_year.to_parquet(f"{GS_dir_output}/全国加油站加气站数据{year}_加油站加气站.parquet")
    print(year, len(GS_data_single_year))
```

```python
encodings_to_try = ['gb2312', 'gbk', 'gb18030', 'utf-8', 'utf-8-sig', 'latin-1']

all_2021_data = None
for encoding in encodings_to_try:
    try:
        print(f"尝试使用编码: {encoding}")
        all_2021_data = pd.read_csv(f"{output_base_path}/加油加气站2021.csv", encoding=encoding)
        print(f"成功使用编码: {encoding}")
        break
    except Exception as e:
        print(f"编码 {encoding} 失败: {str(e)}")
        continue

if all_2021_data is not None:
    print("文件读取成功!")
    all_2021_data
else:
    print("所有编码格式都失败了")
```
