```python
import pandas as pd
```

```python
data_last_2021 = pd.read_csv("C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/raw/全国加油站加气站数据/加油站加气站/加油加气站2021.csv", encoding='gbk')
```

```python
data_last_2021
```

```python
data_last_2021_hn = pd.read_excel("C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/raw/全国加油站加气站数据/加油站加气站/海南.xls")
data_last_2021_hlj = pd.read_excel("C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/raw/全国加油站加气站数据/加油站加气站/黑龙江.xls")
data_last_2021.rename(columns={'wgs84x': 'wgs84_x', 'wgs84y': 'wgs84_y'}, inplace=True)
data_last_2021_hn.rename(columns={'longitude_':'wgs84_x','latitude_g':'wgs84_y'}, inplace=True)
data_last_2021_hlj.rename(columns={'longitude_':'wgs84_x','latitude_g':'wgs84_y'}, inplace=True)
data_last_2021 = pd.concat([data_last_2021, data_last_2021_hn, data_last_2021_hlj], ignore_index=True)
data_last_2021
```

```python
GS_data_new_2021 = pd.read_parquet("C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/intermediate/yiwei/intermidiate/GS_cleaned/GS_data_cleaned_2021.parquet")
GS_data_new_2021
```

```python
CS_data_new_2021 = pd.read_parquet("C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/intermediate/yiwei/intermidiate/CS_cleaned/CS_data_cleaned_2021.parquet")
CS_data_new_2021
```

```python
data_last_2021['cityname'].value_counts()
```

```python
CS_data_new_2021['cityname'].value_counts()
```

```python
matching_columns = ['pname','cityname','adname','address','name','大类','中类','小类']

# Merge dataframes on matching_columns, suffixes to distinguish columns
GS_merged = pd.merge(
    GS_data_new_2021,
    data_last_2021[matching_columns + ['wgs84_x', 'wgs84_y']],
    on=matching_columns,
    how='left',
    suffixes=('', '_last')
)
match_count = GS_merged['wgs84_x_last'].notna().sum()
total_count = len(GS_merged)
match_ratio = match_count / total_count
print(f"Match ratio: {match_ratio:.2%}")
GS_merged
```

```python
GS_merged['wgs84_x'] = GS_merged['wgs84_x_last'].combine_first(GS_merged['wgs84_x'])
GS_merged['wgs84_y'] = GS_merged['wgs84_y_last'].combine_first(GS_merged['wgs84_y'])
```

```python
GS_merged[GS_merged['pname'] == '上海市']
```

```python
GS_merged.drop(columns=['wgs84_x_last', 'wgs84_y_last'], inplace=True)
GS_merged
```

```python
GS_merged.to_parquet("C:/Users/31155/Dropbox/EV-GasDualNetwork/Data/intermediate/yiwei/intermidiate/GS_cleaned/GS_data_cleaned_2021.parquet")
```
